**Impunity Levels**

```python
impunity = df.groupby(['Country Killed', 'Impunity (for Murder)', 'lat', 'lng'])['Name'].count().to_frame().reset_index()
impunity.columns = ['Country Killed', 'Impunity', 'lat', 'lng', 'Name']
impunity
```

```python
impunity['impune'] = impunity['Impunity'].apply(lambda x: 1 if x == 'No' or x == 'Resolved' else 0)
```

```python
impune = impunity.loc[(impunity['impune'] == 0)]
impune = impune.drop('Impunity', axis=1)
impune = impune.groupby(['Country Killed', 'lat', 'lng', 'impune'])['Name'].sum().to_frame().reset_index()
impune
```

```python
fig_impune = px.scatter_mapbox(impune, lat="lat", lon="lng", size="Name",
                    color_continuous_scale=px.colors.cyclical.IceFire_r, 
                    size_max=20, zoom=False, width=600, height=360)
fig_impune.update_geos(
    resolution=50,
    showcoastlines=True, coastlinecolor="RebeccaPurple",
    showland=True, landcolor="LightGreen",
    showocean=True, oceancolor="LightBlue",
    showlakes=True, lakecolor="Blue",
    showrivers=True, rivercolor="Blue"
)
fig_impune.update_layout(height=360, margin={"r":0,"t":0,"l":0,"b":0})
fig_impune.show()
```

```python
top_impunity = impune.sort_values('Name', ascending=False).head(10)
top_impunity
```

```python
graph_impune = alt.Chart(top_impunity).mark_bar(color='orange').encode(
    x=alt.X('Name', axis=alt.Axis(title='Impune Cases')),
    y=alt.Y('Country Killed:N', sort='-x'),
)
graph_impune
```
